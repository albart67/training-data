```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Fixer la graine pour la reproductibilité
np.random.seed(0)

# Nombre de points
m = 50

# Génération des données aléatoires
X = np.random.randn(m).reshape(-1, 1)
y = 1.5 * X + 3 + np.random.randn(m).reshape(-1, 1)

# Créer le modèle de régression linéaire
model = LinearRegression()

# Ajuster le modèle sur les données (X, y)
model.fit(X, y)

# Générer de nouvelles données pour faire des prédictions
new_data = np.linspace(-3, 3, 100).reshape(-1, 1)

# Faire des prédictions avec le modèle ajusté
predictions = model.predict(new_data)

# Tracer les points d'origine
plt.scatter(X, y)

# Tracer la ligne de régression prédite
plt.plot(new_data, predictions, color='red')

# Afficher le graphique
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# Générer des données en clusters avec make_blobs
X, y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=0, cluster_std=1)

# Générer de nouvelles données uniformément réparties dans l'intervalle [X.min(), X.max()]
X_new = np.random.uniform(low=X.min(), high=X.max(), size=(10, 2))

# Afficher les données originales avec des couleurs pour chaque classe
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr')

# Afficher les nouveaux points en blanc avec un contour noir
plt.scatter(X_new[:, 0], X_new[:, 1], c='white', edgecolor='k')

# Afficher le graphique
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Créer un modèle K-Nearest Neighbors avec 3 voisins
model = KNeighborsClassifier(n_neighbors=3)

# Entraîner le modèle sur les données générées (X, y)
model.fit(X, y)
```

```python
# Prédire les étiquettes pour les nouveaux points
predictions = model.predict(X_new)

# Afficher les données originales avec transparence (alpha)
plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.4)

# Afficher les nouveaux points avec les prédictions de couleur et contour noir
plt.scatter(X_new[:, 0], X_new[:, 1], c=predictions, edgecolor='k')

# Afficher le graphique
plt.show()
```

```python

import numpy as np
from sklearn.datasets import make_blobs

# Choisir aléatoirement une valeur de k parmi [2, 3, 4, 5, 6, 7, 8]
k = np.random.choice([2, 3, 4, 5, 6, 7, 8])

# Générer un ensemble de données de 200 échantillons répartis sur k clusters
X, _ = make_blobs(n_samples=200, centers=k, n_features=2, cluster_std=0.5)

# Afficher la valeur de k choisie aléatoirement
print("Valeur de k choisie : ", k)

```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Créer un modèle KMeans avec 4 clusters
model = KMeans(n_clusters=4)

# Ajuster le modèle sur les données X
model.fit(X)

# Prédire les étiquettes des clusters pour chaque point dans X
predictions = model.predict(X)

# Afficher les points avec des couleurs selon leur cluster prédit
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='viridis')

# Afficher le graphique
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixer la graine aléatoire pour la reproductibilité
np.random.seed(0)

# Générer des données aléatoires centrées autour de (0, 0)
X = np.random.randn(30, 2) * 0.2

# Ajouter un point aberrant (outlier) au tableau de données
X = np.concatenate((X, np.array([[2, 0.5]])), axis=0)

# Tracer les données générées
plt.scatter(X[:, 0], X[:, 1])

# Définir les limites des axes
plt.xlim(-2, 3)
plt.ylim(-1, 1)

# Afficher le graphique
plt.show()
```

```python
from sklearn.neighbors import LocalOutlierFactor

# Créer un modèle Local Outlier Factor avec 20 voisins et un taux de contamination de 0.5
model = LocalOutlierFactor(n_neighbors=20, contamination=0.5)

# Ajuster le modèle et prédire les outliers
predictions = model.fit_predict(X)
```

```python
# Afficher les points avec des couleurs en fonction des prédictions (-1 pour outliers, 1 pour normal)
plt.scatter(X[:, 0], X[:, 1], c=predictions, cmap='bwr')

# Définir les limites des axes
plt.xlim(-2, 3)
plt.ylim(-1, 1)

# Afficher le graphique
plt.show()

```
